## 1. 基础语法与Figure对象

### 1.1 导入与基本设置

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 在Jupyter Notebook中显示图表
%matplotlib inline

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
```

**要点解析**：
- `import matplotlib.pyplot as plt`：最常用的导入方式，plt是Matplotlib的顶层接口
- `%matplotlib inline`：Jupyter Notebook专用命令，让图表内联显示
- `plt.rcParams`：全局参数设置，可自定义图表的默认样式

### 1.2 Figure与Axes对象

```python
# 创建Figure和Axes对象
fig, ax = plt.subplots()

# 绘制简单折线图
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y)

# 设置标题和标签
ax.set_title('Simple Sine Wave')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

plt.show()
```

**Figure与Axes的关系**：
- **Figure**：整个图形窗口，相当于画布
- **Axes**：坐标系对象，包含实际的图表内容
- `plt.subplots()`返回Figure对象和Axes对象列表

### 1.3 两种绘图方式

```python
# 方式一：面向对象方式（推荐）
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.show()

# 方式二：pyplot方式（简洁但不灵活）
plt.figure()
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.show()
```

**两种方式对比**：
- 面向对象方式：更适合复杂图表和多子图
- pyplot方式：适合快速简单图表
- 推荐使用面向对象方式，代码更清晰

## 2. 常用图表类型

### 2.1 折线图 (Line Plot)

```python
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y1, label='sin(x)', color='blue', linewidth=2, linestyle='-')
ax.plot(x, y2, label='cos(x)', color='red', linewidth=2, linestyle='--')

ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('Trigonometric Functions')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
```

**常用参数**：
- `label`：图例标签
- `color`或`c`：线条颜色
- `linewidth`或`lw`：线条粗细
- `linestyle`或`ls`：线条样式（'-'实线，'--'虚线，':'点线，'-.'点划线）

### 2.2 散点图 (Scatter Plot)

```python
np.random.seed(42)
n = 100
x = np.random.randn(n)
y = 2 * x + np.random.randn(n) * 0.5
sizes = np.abs(np.random.randn(n)) * 100
colors = np.random.randn(n)

fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')

ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('Scatter Plot with Size and Color')
plt.colorbar(scatter, label='Color Value')

plt.show()
```

**常用参数**：
- `s`：点的大小
- `c`：点的颜色
- `alpha`：透明度（0-1）
- `cmap`：颜色映射表

### 2.3 柱状图 (Bar Plot)

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [23, 45, 56, 78]
values2 = [35, 52, 48, 65]

x = np.arange(len(categories))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, values1, width, label='Group 1', color='steelblue')
bars2 = ax.bar(x + width/2, values2, width, label='Group 2', color='coral')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Chart')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

# 添加数值标签
for bar in bars1:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.show()
```

**常用参数**：
- `width`：柱子宽度
- `label`：图例标签
- `color`：柱子颜色
- `x - width/2`和`x + width/2`：控制分组柱状图的位置

### 2.4 直方图 (Histogram)

```python
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(3, 1, 1000)

fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(data1, bins=30, alpha=0.6, label='Data 1', color='blue')
ax.hist(data2, bins=30, alpha=0.6, label='Data 2', color='red')

ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histogram with Multiple Datasets')
ax.legend()
ax.grid(True, alpha=0.3)

plt.show()
```

**常用参数**：
- `bins`：直方图的柱子数量
- `alpha`：透明度
- `label`：图例标签
- `color`：直方图颜色

### 2.5 饼图 (Pie Chart)

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 35, 20, 20]
explode = (0, 0.1, 0, 0)  # 突出显示第二块
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors,
                                   autopct='%1.1f%%', shadow=True, startangle=90)

for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

ax.set_title('Pie Chart Example')

plt.show()
```

**常用参数**：
- `explode`：突出显示某一块
- `autopct`：显示百分比格式
- `shadow`：添加阴影效果
- `startangle`：起始角度

### 2.6 箱线图 (Box Plot)

```python
np.random.seed(42)
data = [np.random.normal(0, std, 100) for std in range(1, 5)]

fig, ax = plt.subplots(figsize=(10, 6))
bp = ax.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'],
                patch_artist=True)

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax.set_xlabel('Groups')
ax.set_ylabel('Values')
ax.set_title('Box Plot Example')
ax.grid(True, alpha=0.3, axis='y')

plt.show()
```

**箱线图包含的信息**：
- 最小值
- 第一四分位数（Q1）
- 中位数（Q2）
- 第三四分位数（Q3）
- 最大值
- 异常值（outliers）

## 3. 图表定制与美化

### 3.1 设置标题、标签和刻度

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(0, 10, 50)
y = np.sin(x)
ax.plot(x, y)

# 设置标题（支持中文）
ax.set_title('正弦函数图像', fontsize=20, fontweight='bold', pad=20)

# 设置轴标签
ax.set_xlabel('X轴（弧度）', fontsize=14)
ax.set_ylabel('Y轴（正弦值）', fontsize=14)

# 设置刻度标签
ax.tick_params(axis='both', labelsize=12)

# 设置刻度范围
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

plt.show()
```

**常用方法**：
- `set_title()`：设置图表标题
- `set_xlabel()` / `set_ylabel()`：设置轴标签
- `set_xlim()` / `set_ylim()`：设置轴范围
- `tick_params()`：设置刻度样式
- `set_xticks()` / `set_yticks()`：设置刻度位置

### 3.2 添加注释和箭头

```python
fig, ax = plt.subplots(figsize=(10, 6))

x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y)

# 添加注释
ax.annotate('峰值', xy=(np.pi/2, 1), xytext=(np.pi/2+1, 0.8),
            fontsize=12, arrowprops=dict(arrowstyle='->', color='red'))

# 添加文本框
ax.text(6, 0.5, '这是正弦函数', fontsize=12, 
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 添加水平/垂直线
ax.axhline(y=0, color='gray', linestyle='--', linewidth=0.8)
ax.axvline(x=np.pi, color='green', linestyle=':', linewidth=0.8)

plt.show()
```

**常用注释方法**：
- `annotate()`：添加带箭头的注释
- `text()`：添加文本框
- `axhline()` / `axvline()`：添加水平/垂直参考线

### 3.3 设置颜色和样式

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 使用颜色名称
axes[0].plot([1, 2, 3], [1, 2, 3], color='red', label='red')
axes[0].plot([1, 2, 3], [2, 3, 4], color='blue', label='blue')
axes[0].plot([1, 2, 3], [3, 4, 5], color='green', label='green')
axes[0].set_title('Named Colors')
axes[0].legend()

# 使用RGB和RGBA
axes[1].plot([1, 2, 3], [1, 2, 3], color=(0.8, 0.2, 0.2), label='RGB')
axes[1].plot([1, 2, 3], [2, 3, 4], color=(0.2, 0.2, 0.8, 0.5), label='RGBA')
axes[1].plot([1, 2, 3], [3, 4, 5], color='#FF5733', label='Hex')
axes[1].set_title('RGB, RGBA and Hex Colors')
axes[1].legend()

plt.show()
```

**颜色设置方式**：
- 颜色名称：'red', 'blue', 'green'等
- RGB元组：(0.8, 0.2, 0.2)
- RGBA元组：(0.8, 0.2, 0.2, 0.5)
- 十六进制：'#FF5733'

## 4. 多子图与布局

### 4.1 subplots创建多子图

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 第一行第一列
axes[0, 0].plot([1, 2, 3], [1, 2, 3])
axes[0, 0].set_title('Subplot 1')

# 第一行第二列
axes[0, 1].plot([1, 2, 3], [3, 2, 1])
axes[0, 1].set_title('Subplot 2')

# 第二行第一列
axes[1, 0].bar([1, 2, 3], [3, 2, 1])
axes[1, 0].set_title('Subplot 3')

# 第二行第二列
axes[1, 1].scatter([1, 2, 3], [1, 2, 3])
axes[1, 1].set_title('Subplot 4')

fig.suptitle('2x2 Subplots Layout', fontsize=16)
plt.tight_layout()
plt.show()
```

**布局要点**：
- `subplots(2, 2)`：创建2行2列的子图
- `axes[row, col]`：访问特定子图
- `fig.suptitle()`：为整个Figure设置总标题
- `tight_layout()`：自动调整子图间距

### 4.2 使用subplot创建不规则布局

```python
fig = plt.figure(figsize=(12, 8))

# 创建不规则布局
ax1 = plt.subplot(2, 2, 1)  # 第一行第一列
ax2 = plt.subplot(2, 2, 2)  # 第一行第二列
ax3 = plt.subplot(2, 1, 2)  # 占满整个第二行

ax1.plot([1, 2, 3], [1, 2, 3])
ax1.set_title('Plot 1')

ax2.plot([1, 2, 3], [3, 2, 1])
ax2.set_title('Plot 2')

ax3.bar([1, 2, 3], [5, 3, 7])
ax3.set_title('Plot 3 (Full Width)')

plt.tight_layout()
plt.show()
```

**subplot参数**：
- 第一个参数：行数
- 第二个参数：列数
- 第三个参数：子图编号（从1开始）

### 4.3 GridSpec自定义布局

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(12, 8))
gs = gridspec.GridSpec(3, 3, figure=fig)

ax1 = fig.add_subplot(gs[0, :])  # 第一行，占满所有列
ax2 = fig.add_subplot(gs[1, :-1])  # 第二行，前两列
ax3 = fig.add_subplot(gs[1:, -1])  # 右侧所有行
ax4 = fig.add_subplot(gs[-1, 0])  # 最后一行，第一列
ax5 = fig.add_subplot(gs[-1, -2])  # 最后一行，倒数第二列

ax1.plot([1, 2, 3], [1, 2, 3])
ax1.set_title('Full Width')

ax2.bar([1, 2, 3], [3, 2, 1])
ax2.set_title('Partial Width')

ax3.scatter([1, 2, 3], [1, 2, 3])
ax3.set_title('Full Height')

ax4.text(0.5, 0.5, 'Text Box 1', ha='center', va='center')
ax5.text(0.5, 0.5, 'Text Box 2', ha='center', va='center')

plt.tight_layout()
plt.show()
```

**GridSpec切片语法**：
- `gs[0, :]`：第一行，所有列
- `gs[1:, -1]`：最后一列，所有行
- `gs[-1, 0]`：最后一行，第一列

## 5. 高级技巧

### 5.1 保存图表

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot([1, 2, 3], [1, 4, 2])
ax.set_title('Save Figure Example')

# 保存为不同格式
fig.savefig('figure.png', dpi=300, bbox_inches='tight')  # PNG格式
fig.savefig('figure.pdf', bbox_inches='tight')  # PDF格式
fig.savefig('figure.svg', bbox_inches='tight')  # SVG格式

print('图表已保存为PNG、PDF和SVG格式')
```

**保存参数**：
- `dpi`：分辨率（默认100，出版品质通常用300）
- `bbox_inches='tight'`：紧凑裁剪白边
- 支持格式：PNG、PDF、SVG、EPS等

### 5.2 双Y轴图表

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.exp(x)

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('X values')
ax1.set_ylabel('sin(x)', color=color)
ax1.plot(x, y1, color=color, label='sin(x)')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # 创建共享X轴的第二个Y轴
color = 'tab:red'
ax2.set_ylabel('exp(x)', color=color)
ax2.plot(x, y2, color=color, label='exp(x)')
ax2.tick_params(axis='y', labelcolor=color)

fig.suptitle('Dual Y-Axis Plot')
plt.show()
```

**双Y轴要点**：
- `twinx()`：创建共享X轴的第二个Y轴
- 常用于显示不同量级的数据

### 5.3 填充区域

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 填充曲线下方区域
axes[0].fill_between(x, y, alpha=0.3)
axes[0].plot(x, y)
axes[0].set_title('Fill Below')

# 在两条曲线之间填充
y2 = np.cos(x)
axes[1].fill_between(x, y, y2, alpha=0.3, color='green')
axes[1].plot(x, y, label='sin(x)')
axes[1].plot(x, y2, label='cos(x)')
axes[1].set_title('Fill Between Curves')
axes[1].legend()

plt.show()
```

**填充函数**：
- `fill_between(x, y)`：填充曲线下方
- `fill_between(x, y1, y2)`：在两条曲线之间填充

### 5.4 添加颜色条(Colorbar)

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(Z, cmap='RdYlBu_r', extent=[-3, 3, -3, 3], origin='lower')

cbar = plt.colorbar(im, ax=ax)
cbar.set_label('Value', fontsize=12)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('2D Color Plot with Colorbar')

plt.show()
```

**Colorbar要点**：
- `imshow()`：显示2D数组
- `plt.colorbar()`：添加颜色条
- `cmap`：选择颜色映射

### 5.5 设置对数坐标

```python
x = np.linspace(1, 100, 100)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 线性坐标
axes[0].plot(x, y)
axes[0].set_title('Linear Scale')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

# 对数坐标
axes[1].plot(x, y)
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_title('Log Scale')
axes[1].set_xlabel('X (log)')
axes[1].set_ylabel('Y (log)')

plt.tight_layout()
plt.show()
```

**对数坐标**：
- `set_xscale('log')`：X轴使用对数刻度
- `set_yscale('log')`：Y轴使用对数刻度
- 适用于跨多个数量级的数据
