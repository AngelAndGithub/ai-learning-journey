# 数据分布分析详解

## 简介

数据分布分析是数据分析的重要环节，它帮助我们了解数据的基本特征，包括数据的集中趋势、离散程度、分布形态等。通过数据分布分析，我们可以：

1. **识别数据异常**：发现数据中的异常值或离群点
2. **了解数据特征**：掌握各变量的分布规律
3. **指导后续分析**：为特征工程和建模提供依据

## 主要部分

### 1. 数值型变量分布分析

**功能解析**：
- **选择数值型变量**：使用`select_dtypes`方法筛选出所有数值型列（int64和float64类型）
- **绘制直方图**：对每个数值型变量绘制直方图，展示数据的分布情况
- **添加KDE曲线**：`kde=True`参数添加核密度估计曲线，更直观地展示数据分布形态
- **保存图表**：将每个变量的分布图保存为单独的图片文件

```python
# 数值型变量分布分析代码示例
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 获取当前目录
current_dir = os.path.dirname(os.path.abspath(__file__))

# 读取数据
df = pd.read_csv(os.path.join(current_dir, 'customer_data.csv'))

# 选择数值型变量
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 绘制数值型变量分布图
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True)
    plt.title(f'{col} 分布')
    plot_path = os.path.join(current_dir, f'{col}_distribution.png')
    plt.savefig(plot_path)
    plt.close()

print("数值型变量分布图已保存")
```

**直方图的作用**：
- 展示数据在不同区间的频率分布
- 帮助识别数据的集中趋势（如均值、中位数）
- 观察数据的离散程度
- 识别数据是否符合正态分布

### 2. 分类型变量分布分析

**功能解析**：
- **选择分类型变量**：使用`select_dtypes`方法筛选出所有分类型列（object类型）
- **绘制计数图**：对每个分类型变量绘制计数图，展示不同类别的数量分布
- **旋转x轴标签**：`plt.xticks(rotation=45)`旋转x轴标签，避免类别名称过长时重叠
- **保存图表**：将每个变量的分布图保存为单独的图片文件

```python
# 分类型变量分布分析代码示例
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 获取当前目录
current_dir = os.path.dirname(os.path.abspath(__file__))

# 读取数据
df = pd.read_csv(os.path.join(current_dir, 'customer_data.csv'))

# 选择分类型变量
categorical_cols = df.select_dtypes(include=['object']).columns

# 绘制分类型变量分布图
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df)
    plt.title(f'{col} 分布')
    plt.xticks(rotation=45)
    plot_path = os.path.join(current_dir, f'{col}_distribution.png')
    plt.savefig(plot_path)
    plt.close()

print("分类型变量分布图已保存")
```

**计数图的作用**：
- 展示不同类别的数量分布
- 识别数据中的类别不平衡问题
- 了解类别之间的比例关系

## 分析结果的意义

1. **异常值检测**：通过观察直方图，我们可以发现数据中的异常值（如远离大部分数据的极端值）
2. **数据预处理指导**：根据分布情况，我们可以决定是否需要对数据进行标准化、归一化或其他预处理
3. **特征工程依据**：了解数据分布后，我们可以更有针对性地创建新特征
4. **模型选择参考**：不同的模型对数据分布有不同的假设，了解数据分布有助于选择合适的模型

## 示例解释

假设我们有以下变量：
- **数值型变量**：age（年龄）、income（收入）、spending_score（支出得分）
- **分类型变量**：gender（性别）、product_category（产品类别）、region（地区）

通过数据分布分析，我们可以：
- 了解年龄的分布情况（是否集中在某个年龄段）
- 观察收入的分布是否符合正态分布
- 分析不同性别的分布比例
- 了解不同产品类别的受欢迎程度
- 掌握不同地区的客户分布情况

这些信息对于后续的客户分群、营销策略制定等都有重要的参考价值。

## 代码优化建议

1. **添加分布统计信息**：可以在绘制图表的同时，计算并打印每个变量的分布统计信息（如均值、中位数、标准差等）
2. **增加分布检验**：可以添加正态分布检验，判断数据是否符合正态分布
3. **优化图表样式**：可以添加网格线、调整颜色等，使图表更加美观
4. **添加交互功能**：如果在Jupyter Notebook中运行，可以添加交互功能，让用户可以更直观地探索数据分布

```python
# 代码优化示例：添加分布统计信息
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 获取当前目录
current_dir = os.path.dirname(os.path.abspath(__file__))

# 读取数据
df = pd.read_csv(os.path.join(current_dir, 'customer_data.csv'))

# 选择数值型变量
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 绘制数值型变量分布图并添加统计信息
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True)
    
    # 计算统计信息
    mean_val = df[col].mean()
    median_val = df[col].median()
    std_val = df[col].std()
    
    # 添加统计信息到图表
    plt.axvline(mean_val, color='r', linestyle='--', label=f'均值: {mean_val:.2f}')
    plt.axvline(median_val, color='g', linestyle='--', label=f'中位数: {median_val:.2f}')
    plt.legend()
    
    plt.title(f'{col} 分布 (标准差: {std_val:.2f})')
    plot_path = os.path.join(current_dir, f'{col}_distribution.png')
    plt.savefig(plot_path)
    plt.close()
    
    # 打印统计信息
    print(f'{col} 统计信息:')
    print(f'  均值: {mean_val:.2f}')
    print(f'  中位数: {median_val:.2f}')
    print(f'  标准差: {std_val:.2f}')
    print(f'  最小值: {df[col].min():.2f}')
    print(f'  最大值: {df[col].max():.2f}')
    print()

print("数值型变量分布图已保存")
```
