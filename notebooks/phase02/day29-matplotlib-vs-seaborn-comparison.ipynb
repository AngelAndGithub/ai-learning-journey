# Matplotlib 与 Seaborn 区别与使用详解

## 简介

Matplotlib 和 Seaborn 是 Python 中最流行的数据可视化库，它们各有特点和适用场景。本教程将详细对比两者的区别，并提供实际使用指南。

**Matplotlib**：
- Python 数据可视化的基础库
- 提供底层绘图功能
- 高度可定制
- 适合创建复杂的图表

**Seaborn**：
- 基于 Matplotlib 构建
- 提供更高级的接口
- 美观的默认样式
- 专为统计数据可视化设计

## 1. 核心区别对比

| 特性 | Matplotlib | Seaborn |
|------|------------|---------|
| **基础定位** | 底层绘图库 | 高级统计可视化库 |
| **依赖关系** | 独立库 | 基于 Matplotlib |
| **默认样式** | 基础、简洁 | 美观、现代化 |
| **API 风格** | 命令式 + 面向对象 | 声明式、简洁 |
| **统计功能** | 基础 | 强大、内置统计图表 |
| **复杂度** | 高（需要更多代码） | 低（简洁的 API） |
| **适用场景** | 复杂定制、底层控制 | 统计分析、快速可视化 |
| **学习曲线** | 较陡峭 | 较平缓 |
| **性能** | 通常更快 | 可能稍慢（封装了更多功能） |

## 3. 基础绘图对比

### 3.1 折线图对比

```python
# 生成数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Matplotlib 折线图
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Matplotlib 版本
ax[0].plot(x, y1, label='sin(x)', color='blue', linewidth=2)
ax[0].plot(x, y2, label='cos(x)', color='red', linewidth=2, linestyle='--')
ax[0].set_title('Matplotlib 折线图')
ax[0].set_xlabel('X 轴')
ax[0].set_ylabel('Y 轴')
ax[0].legend()
ax[0].grid(True, alpha=0.3)

# Seaborn 版本
sns.lineplot(x=x, y=y1, label='sin(x)', ax=ax[1])
sns.lineplot(x=x, y=y2, label='cos(x)', ax=ax[1])
ax[1].set_title('Seaborn 折线图')
ax[1].set_xlabel('X 轴')
ax[1].set_ylabel('Y 轴')

plt.tight_layout()
plt.show()
```

**对比分析**：
- Matplotlib：需要更多代码设置标题、标签和图例
- Seaborn：自动处理图例，默认样式更美观

### 3.2 散点图对比

```python
# 生成数据
np.random.seed(42)
n = 100
x = np.random.randn(n)
y = 2 * x + np.random.randn(n) * 0.5
colors = np.random.randn(n)
sizes = np.abs(np.random.randn(n)) * 100

# 对比散点图
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Matplotlib 版本
scatter1 = ax[0].scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
ax[0].set_title('Matplotlib 散点图')
ax[0].set_xlabel('X 轴')
ax[0].set_ylabel('Y 轴')
plt.colorbar(scatter1, ax=ax[0])

# Seaborn 版本
sns.scatterplot(x=x, y=y, size=sizes, hue=colors, alpha=0.6, palette='viridis', ax=ax[1])
ax[1].set_title('Seaborn 散点图')
ax[1].set_xlabel('X 轴')
ax[1].set_ylabel('Y 轴')

plt.tight_layout()
plt.show()
```

**对比分析**：
- Matplotlib：需要手动添加颜色条，参数较多
- Seaborn：自动处理颜色和大小，API 更简洁

## 4. 样式与美观度

### 4.1 样式对比

```python
# 样式对比
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Matplotlib 默认样式
ax[0].plot(x, y1, label='sin(x)')
ax[0].plot(x, y2, label='cos(x)')
ax[0].set_title('Matplotlib 默认样式')
ax[0].legend()

# Seaborn 默认样式
with sns.axes_style('darkgrid'):
    sns.lineplot(x=x, y=y1, label='sin(x)', ax=ax[1])
    sns.lineplot(x=x, y=y2, label='cos(x)', ax=ax[1])
    ax[1].set_title('Seaborn 默认样式')

plt.tight_layout()
plt.show()
```

### 4.2 样式设置

```python
# 查看 Seaborn 可用样式
print('Seaborn 可用样式:', sns.axes_style())
print('
Seaborn 可用主题:', sns.get_dataset_names())

# 对比不同样式
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, style in enumerate(styles):
    with sns.axes_style(style):
        ax = axes[i]
        sns.lineplot(x=x, y=y1, ax=ax)
        ax.set_title(f'Style: {style}')

plt.tight_layout()
plt.show()
```

**样式优势**：
- Matplotlib：高度可定制，但需要更多代码
- Seaborn：内置多种美观样式，一行代码即可切换

## 5. 统计功能对比

### 5.1 直方图对比

```python
# 加载示例数据
tips = sns.load_dataset('tips')

# 直方图对比
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Matplotlib 版本
ax[0].hist(tips['total_bill'], bins=30, alpha=0.6, color='blue')
ax[0].set_title('Matplotlib 直方图')
ax[0].set_xlabel('总消费')
ax[0].set_ylabel('频率')

# Seaborn 版本
sns.histplot(data=tips, x='total_bill', bins=30, kde=True, ax=ax[1])
ax[1].set_title('Seaborn 直方图')
ax[1].set_xlabel('总消费')
ax[1].set_ylabel('频率')

plt.tight_layout()
plt.show()
```

### 5.2 箱线图对比

```python
# 箱线图对比
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Matplotlib 版本
ax[0].boxplot([tips[tips['day'] == day]['total_bill'] for day in tips['day'].unique()])
ax[0].set_title('Matplotlib 箱线图')
ax[0].set_xlabel('星期')
ax[0].set_ylabel('总消费')
ax[0].set_xticklabels(tips['day'].unique())

# Seaborn 版本
sns.boxplot(data=tips, x='day', y='total_bill', ax=ax[1])
ax[1].set_title('Seaborn 箱线图')
ax[1].set_xlabel('星期')
ax[1].set_ylabel('总消费')

plt.tight_layout()
plt.show()
```

### 5.3 线性回归对比

```python
# 线性回归对比
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Matplotlib 版本（需要手动计算回归）
from sklearn.linear_model import LinearRegression
X = tips['total_bill'].values.reshape(-1, 1)
y = tips['tip'].values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

ax[0].scatter(tips['total_bill'], tips['tip'])
ax[0].plot(tips['total_bill'], y_pred, color='red')
ax[0].set_title('Matplotlib 线性回归')
ax[0].set_xlabel('总消费')
ax[0].set_ylabel('小费')

# Seaborn 版本
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax[1])
ax[1].set_title('Seaborn 线性回归')
ax[1].set_xlabel('总消费')
ax[1].set_ylabel('小费')

plt.tight_layout()
plt.show()
```

**统计功能对比**：
- Matplotlib：需要手动实现统计功能
- Seaborn：内置丰富的统计图表和功能

## 6. 适用场景分析

### 6.1 Matplotlib 适用场景

```python
# Matplotlib 适用场景：复杂定制
fig = plt.figure(figsize=(12, 8))

# 创建极坐标图
ax1 = fig.add_subplot(221, projection='polar')
theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(3*theta)
ax1.plot(theta, r)
ax1.set_title('极坐标图')

# 创建3D图
from mpl_toolkits.mplot3d import Axes3D
ax2 = fig.add_subplot(222, projection='3d')
X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = np.sin(np.sqrt(X**2 + Y**2))
ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_title('3D表面图')

# 创建饼图
ax3 = fig.add_subplot(223)
sizes = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']
ax3.pie(sizes, labels=labels, autopct='%1.1f%%')
ax3.set_title('饼图')

# 创建误差线图
ax4 = fig.add_subplot(224)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
y_err = [0.5, 0.8, 0.3, 0.9, 0.4]
ax4.errorbar(x, y, yerr=y_err, fmt='o-', capsize=5)
ax4.set_title('误差线图')

plt.tight_layout()
plt.show()
```

**Matplotlib 适用场景**：
- 复杂的图表类型（极坐标图、3D图等）
- 需要完全控制图表的每一个元素
- 非统计类数据可视化
- 需要高度定制的出版级图表

### 6.2 Seaborn 适用场景

```python
# Seaborn 适用场景：统计分析
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 联合分布图
sns.jointplot(data=tips, x='total_bill', y='tip', ax=axes[0, 0])
axes[0, 0].set_title('联合分布图')

# 成对关系图
sns.pairplot(data=tips, vars=['total_bill', 'tip', 'size'], hue='time', ax=axes[0, 1])
axes[0, 1].set_title('成对关系图')

# 热图
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[1, 0])
axes[1, 0].set_title('相关性热图')

# 分类小提琴图
sns.violinplot(data=tips, x='day', y='total_bill', hue='time', split=True, ax=axes[1, 1])
axes[1, 1].set_title('分类小提琴图')

plt.tight_layout()
plt.show()
```

**Seaborn 适用场景**：
- 统计数据可视化
- 快速探索性数据分析
- 美观的默认样式
- 复杂的多变量分析
- 分类数据的比较

## 7. 如何结合使用

```python
# 结合使用示例
fig, ax = plt.subplots(figsize=(12, 6))

# 使用 Seaborn 创建基础图表
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size', ax=ax)

# 使用 Matplotlib 进行精细定制
ax.set_title('消费与小费关系分析', fontsize=16, fontweight='bold')
ax.set_xlabel('总消费金额', fontsize=12)
ax.set_ylabel('小费金额', fontsize=12)

# 添加自定义注释
mean_total_bill = tips['total_bill'].mean()
mean_tip = tips['tip'].mean()
ax.axvline(mean_total_bill, color='red', linestyle='--', label=f'平均消费: {mean_total_bill:.2f}')
ax.axhline(mean_tip, color='green', linestyle='--', label=f'平均小费: {mean_tip:.2f}')

# 添加文本框
ax.text(40, 1, '数据来源: tips 数据集', ha='right', va='bottom',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 自定义图例
legend = ax.legend(loc='upper left', fontsize=10)
legend.get_title().set_fontsize(12)

# 添加网格
ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```

**结合使用的优势**：
1. **效率与美观**：使用 Seaborn 快速创建美观的基础图表
2. **精细定制**：使用 Matplotlib 进行底层定制
3. **功能互补**：Seaborn 提供统计功能，Matplotlib 提供底层控制
4. **灵活性**：根据具体需求选择合适的工具

## 8. 性能与效率

```python
# 性能对比测试
import time

# 生成大型数据集
np.random.seed(42)
n = 100000
large_data = pd.DataFrame({
    'x': np.random.randn(n),
    'y': np.random.randn(n),
    'category': np.random.choice(['A', 'B', 'C'], n)
})

# 测试 Matplotlib 性能
start_time = time.time()
fig, ax = plt.subplots()
ax.scatter(large_data['x'], large_data['y'], alpha=0.1)
plt.close()  # 不显示，只测试绘制时间
matplotlib_time = time.time() - start_time
print(f'Matplotlib 散点图绘制时间: {matplotlib_time:.4f} 秒')

# 测试 Seaborn 性能
start_time = time.time()
fig, ax = plt.subplots()
sns.scatterplot(data=large_data, x='x', y='y', alpha=0.1, ax=ax)
plt.close()  # 不显示，只测试绘制时间
seaborn_time = time.time() - start_time
print(f'Seaborn 散点图绘制时间: {seaborn_time:.4f} 秒')

# 计算性能差异
ratio = seaborn_time / matplotlib_time
print(f'Seaborn 比 Matplotlib 慢 {ratio:.2f} 倍')
```

**性能分析**：
- Matplotlib：通常更快，尤其是处理大型数据集时
- Seaborn：由于额外的统计计算和样式处理，可能稍慢
- 对于大多数日常使用场景，性能差异可以忽略不计

## 总结

### 核心对比总结

| 方面 | Matplotlib | Seaborn |
|------|------------|---------|
| **定位** | 底层绘图库 | 高级统计可视化库 |
| **优势** | 高度可定制、灵活、性能好 | 简洁API、美观默认样式、强大统计功能 |
| **劣势** | 代码量大、默认样式一般 | 性能稍慢、底层控制能力弱 |
| **最佳用途** | 复杂定制图表、特殊图表类型 | 统计分析、数据探索、快速可视化 |
| **学习曲线** | 较陡峭 | 较平缓 |
| **推荐场景** | 需要完全控制图表细节时 | 统计数据分析和快速可视化时 |

### 最终建议

1. **初学者**：先学习 Seaborn，快速上手并创建美观的图表
2. **进阶用户**：掌握 Matplotlib 以实现更复杂的定制
3. **专业用户**：结合使用两者，发挥各自优势
4. **实际项目**：根据具体需求选择合适的工具，或结合使用

Matplotlib 和 Seaborn 不是竞争关系，而是互补关系。它们共同构成了 Python 数据可视化的强大生态系统，为不同场景提供了合适的解决方案。
