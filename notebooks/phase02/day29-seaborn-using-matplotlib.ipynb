## 1. Seaborn 与 Matplotlib 的关系

```python
# 查看 Seaborn 的依赖关系
import seaborn as sns
import matplotlib

print(f'Seaborn 版本: {sns.__version__}')
print(f'Matplotlib 版本: {matplotlib.__version__}')
print(f'Seaborn 是否使用 Matplotlib: {sns._core.plotting_context is not None}')
```

**关系说明**：
1. **依赖关系**：Seaborn 依赖 Matplotlib 作为底层绘图引擎
2. **对象关系**：Seaborn 创建的图表实际上是 Matplotlib 的 Figure 和 Axes 对象
3. **API 层次**：Seaborn 提供更高级的 API，而 Matplotlib 提供底层控制

## 2. Seaborn 如何使用 Matplotlib 对象

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 加载示例数据
tips = sns.load_dataset('tips')

# 示例1：Seaborn 内部使用 Matplotlib 的 Figure 和 Axes
print('=== 示例1：Seaborn 图表的类型 ===')

# 创建 Seaborn 图表
ax = sns.scatterplot(data=tips, x='total_bill', y='tip')

# 查看返回对象的类型
print(f'Seaborn scatterplot 返回类型: {type(ax)}')
print(f'是 Matplotlib Axes 对象: {isinstance(ax, plt.Axes)}')

# 获取 Figure 对象
fig = ax.get_figure()
print(f'Figure 对象类型: {type(fig)}')
print(f'是 Matplotlib Figure 对象: {isinstance(fig, plt.Figure)}')

plt.close()  # 关闭图表

# 示例2：Seaborn 如何创建 Figure
print('=== 示例2：Seaborn 创建 Figure ===')

# 直接创建 Figure 和 Axes
fig, ax = plt.subplots()
print(f'手动创建的 Axes: {type(ax)}')

# 在指定的 Axes 上绘制 Seaborn 图表
sns.lineplot(data=tips, x='total_bill', y='tip', ax=ax)
print(f'Seaborn 在指定 Axes 上绘制: 成功')

plt.close()
```

**关键发现**：
- Seaborn 函数返回的是 Matplotlib 的 Axes 对象
- Seaborn 可以在手动创建的 Matplotlib Axes 上绘制
- 所有 Seaborn 图表都是基于 Matplotlib 的 Figure 和 Axes

---

## 3. 在 Seaborn 图表上使用 Matplotlib 方法

```python
# 示例：在 Seaborn 图表上使用 Matplotlib 方法
fig, ax = plt.subplots(figsize=(10, 6))

# 使用 Seaborn 创建基础图表
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', ax=ax)

# 使用 Matplotlib 方法进行定制
# 1. 设置标题
ax.set_title('消费与小费关系', fontsize=16, fontweight='bold')

# 2. 设置轴标签
ax.set_xlabel('总消费金额 (元)', fontsize=12)
ax.set_ylabel('小费金额 (元)', fontsize=12)

# 3. 添加参考线
mean_total_bill = tips['total_bill'].mean()
mean_tip = tips['tip'].mean()
ax.axvline(mean_total_bill, color='red', linestyle='--', label=f'平均消费: {mean_total_bill:.2f}')
ax.axhline(mean_tip, color='green', linestyle='--', label=f'平均小费: {mean_tip:.2f}')

# 4. 添加文本注释
ax.text(40, 1, '数据来源: tips 数据集', ha='right', va='bottom',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# 5. 自定义图例
legend = ax.legend(loc='upper left', fontsize=10)
legend.get_title().set_fontsize(12)

# 6. 添加网格
ax.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```

**使用方法**：
1. **获取 Axes 对象**：Seaborn 函数返回的是 Matplotlib Axes 对象
2. **直接调用方法**：在该 Axes 对象上直接调用 Matplotlib 的方法
3. **所有 Matplotlib 方法都可用**：包括 set_title、set_xlabel、axvline 等

## 4. 将 Seaborn 图表添加到 Matplotlib 子图中

```python
# 示例：在 Matplotlib 子图布局中使用 Seaborn
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 第一个子图：Seaborn 散点图
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', ax=axes[0, 0])
axes[0, 0].set_title('散点图')

# 第二个子图：Seaborn 箱线图
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0, 1])
axes[0, 1].set_title('箱线图')

# 第三个子图：Seaborn 直方图
sns.histplot(data=tips, x='total_bill', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('直方图')

# 第四个子图：Matplotlib 饼图
day_counts = tips['day'].value_counts()
axes[1, 1].pie(day_counts, labels=day_counts.index, autopct='%1.1f%%')
axes[1, 1].set_title('饼图')

# 整体标题
fig.suptitle('Seaborn 与 Matplotlib 混合使用示例', fontsize=16, y=1.02)

plt.tight_layout()
plt.show()
```

**布局技巧**：
1. **创建子图布局**：使用 `plt.subplots()` 创建网格布局
2. **指定 ax 参数**：在 Seaborn 函数中使用 `ax` 参数指定子图
3. **混合使用**：在不同子图中混合使用 Seaborn 和 Matplotlib

## 5. 混合使用 Seaborn 和 Matplotlib

```python
# 高级示例：混合使用 Seaborn 和 Matplotlib
fig, ax = plt.subplots(figsize=(12, 7))

# 1. 使用 Seaborn 创建基础图表
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax,
            scatter_kws={'alpha': 0.6, 's': 50},
            line_kws={'color': 'red', 'linewidth': 2})

# 2. 使用 Matplotlib 添加额外元素
# 添加数据点标签
for i, (x, y) in enumerate(zip(tips['total_bill'][:10], tips['tip'][:10])):
    ax.annotate(f'#{i+1}', xy=(x, y), xytext=(5, 5),
                textcoords='offset points', fontsize=8,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

# 3. 添加自定义区域
ax.axvspan(30, 50, alpha=0.1, color='yellow', label='高消费区域')
ax.axhspan(5, 10, alpha=0.1, color='green', label='高小费区域')

# 4. 自定义坐标轴
ax.set_xlim(0, 60)
ax.set_ylim(0, 12)
ax.set_xticks(np.arange(0, 61, 10))
ax.set_yticks(np.arange(0, 13, 2))

# 5. 美化图表
ax.set_title('消费与小费关系分析', fontsize=16, fontweight='bold')
ax.set_xlabel('总消费金额', fontsize=12)
ax.set_ylabel('小费金额', fontsize=12)
ax.legend(loc='upper left')
ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.show()
```

**混合使用的优势**：
1. **快速创建**：使用 Seaborn 快速创建美观的基础图表
2. **精细定制**：使用 Matplotlib 进行底层定制
3. **功能互补**：结合两者的优势
4. **灵活性**：根据需要选择合适的工具

## 6. 常见问题解答

### 6.1 为什么在 Seaborn 代码中看不到 Matplotlib 的使用？

```python
# 示例：Seaborn 隐藏了 Matplotlib 的调用
print('=== Seaborn 如何隐藏 Matplotlib 调用 ===')

# 简单的 Seaborn 代码
import seaborn as sns
import matplotlib.pyplot as plt

# 这段代码看起来没有使用 Matplotlib
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()

print('
但实际上：')
print('1. Seaborn 在内部创建了 Matplotlib Figure 和 Axes')
print('2. Seaborn 使用 Matplotlib 的绘图函数')
print('3. plt.show() 是 Matplotlib 的函数')
```

### 6.2 如何确认 Seaborn 使用的是 Matplotlib？

```python
# 验证 Seaborn 使用 Matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# 创建 Seaborn 图表
ax = sns.lineplot(data=tips, x='total_bill', y='tip')

# 验证对象类型
print(f'Seaborn 返回的对象类型: {type(ax)}')
print(f'是否为 Matplotlib Axes: {isinstance(ax, plt.Axes)}')

# 验证 Figure 对象
fig = ax.get_figure()
print(f'
Figure 对象类型: {type(fig)}')
print(f'是否为 Matplotlib Figure: {isinstance(fig, plt.Figure)}')

# 查看 Figure 中的所有 Axes
print(f'
Figure 中的 Axes 数量: {len(fig.axes)}')

plt.close()
```

### 6.3 Seaborn 可以完全替代 Matplotlib 吗？

```python
# 示例：Matplotlib 能做但 Seaborn 不能直接做的事情
import matplotlib.pyplot as plt
import numpy as np

print('=== Matplotlib 特有功能 ===')

# 1. 极坐标图
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(3*theta)
ax.plot(theta, r)
ax.set_title('极坐标图')
plt.close()
print('1. 极坐标图: Matplotlib 支持，Seaborn 不直接支持')

# 2. 3D 图表
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.close()
print('2. 3D 图表: Matplotlib 支持，Seaborn 不直接支持')

# 3. 自定义形状和路径
fig, ax = plt.subplots()
from matplotlib.patches import Circle, Rectangle
circle = Circle((0.5, 0.5), 0.3, color='blue', alpha=0.5)
rect = Rectangle((0.1, 0.1), 0.3, 0.3, color='red', alpha=0.5)
ax.add_patch(circle)
ax.add_patch(rect)
plt.close()
print('3. 自定义形状: Matplotlib 支持，Seaborn 不直接支持')
```

## 总结

### Seaborn 与 Matplotlib 的关系

1. **底层依赖**：Seaborn 完全依赖 Matplotlib 作为底层绘图引擎
2. **对象共享**：Seaborn 创建的图表是 Matplotlib 的 Figure 和 Axes 对象
3. **无缝集成**：可以在 Seaborn 图表上直接使用 Matplotlib 方法
4. **功能互补**：Seaborn 提供高级统计图表，Matplotlib 提供底层控制

### 如何在 Seaborn 中使用 Matplotlib

1. **获取 Axes 对象**：Seaborn 函数返回 Matplotlib Axes 对象
2. **直接调用方法**：在该对象上调用 Matplotlib 的各种方法
3. **指定 ax 参数**：在 Seaborn 函数中指定 Matplotlib Axes
4. **混合使用**：在同一图表中混合使用两者的功能

### 最佳实践

1. **快速可视化**：使用 Seaborn 的高级 API
2. **精细定制**：使用 Matplotlib 进行底层调整
3. **复杂布局**：使用 Matplotlib 创建布局，Seaborn 填充内容
4. **特殊图表**：使用 Matplotlib 创建 Seaborn 不支持的图表类型

Seaborn 和 Matplotlib 不是替代关系，而是互补关系。Seaborn 简化了统计数据可视化的过程，而 Matplotlib 提供了最终的定制能力。通过理解它们之间的关系，你可以充分发挥两者的优势，创建既美观又功能强大的数据可视化。
