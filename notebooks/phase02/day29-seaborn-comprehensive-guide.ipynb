# Seaborn 知识点详解

## 简介

Seaborn是基于Matplotlib的Python数据可视化库，提供了更高级的接口和更美观的默认样式。它专为统计数据可视化而设计，使复杂的数据可视化变得简单。

**主要特点**：
- 美观的默认样式
- 专门的统计图表类型
- 强大的数据集处理能力
- 与Pandas无缝集成
- 支持复杂的多变量可视化

## 1. 基础设置与样式

### 1.1 导入与基本设置

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 在Jupyter Notebook中显示图表
%matplotlib inline

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 查看可用的Seaborn样式
print('可用样式:', sns.get_dataset_names())
```

**Seaborn数据集**：Seaborn内置了多个示例数据集，可以通过`load_dataset()`加载

### 1.2 样式设置

```python
# 设置默认样式
sns.set_style('darkgrid')  # 深色网格

# 可用样式：darkgrid, whitegrid, dark, white, ticks

# 设置字体大小
sns.set_context('notebook', font_scale=1.2)

# 可用上下文：paper, notebook, talk, poster

# 设置颜色主题
sns.set_palette('viridis')  # 使用viridis调色板

# 演示不同样式
styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, style in enumerate(styles):
    with sns.axes_style(style):
        ax = axes[i]
        sns.lineplot(x=[1, 2, 3, 4], y=[2, 4, 1, 3], ax=ax)
        ax.set_title(f'Style: {style}')

plt.tight_layout()
plt.show()
```

**样式设置方法**：
- `set_style()`：设置整体样式
- `set_context()`：设置字体大小和线条粗细
- `set_palette()`：设置颜色主题
- `axes_style()`：局部样式设置（使用with语句）

### 1.3 颜色主题

```python
# 查看可用调色板
palettes = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 'coolwarm', 'RdBu', 'Set2']

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, palette in enumerate(palettes):
    ax = axes[i]
    sns.palplot(sns.color_palette(palette), ax=ax)
    ax.set_title(palette)

plt.tight_layout()
plt.show()
```

**调色板类型**：
- 顺序调色板：viridis, plasma, inferno等
- 发散调色板：coolwarm, RdBu等
- 分类调色板：Set2, tab10等

## 2. 单变量分布可视化

### 2.1 直方图与核密度估计

```python
# 加载示例数据
tips = sns.load_dataset('tips')

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 简单直方图
sns.histplot(data=tips, x='total_bill', ax=axes[0])
axes[0].set_title('简单直方图')

# 带核密度估计的直方图
sns.histplot(data=tips, x='total_bill', kde=True, ax=axes[1])
axes[1].set_title('带KDE的直方图')

# 单独的核密度估计图
sns.kdeplot(data=tips, x='total_bill', ax=axes[2])
axes[2].set_title('核密度估计图')

plt.tight_layout()
plt.show()
```

**histplot参数**：
- `data`：数据集
- `x`：要绘制的变量
- `kde`：是否显示核密度估计
- `bins`：直方图的柱数
- `color`：颜色

### 2.2 分布组合图

```python
# 使用displot创建分布组合图
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 直方图+KDE
sns.displot(data=tips, x='total_bill', kde=True, ax=axes[0])
axes[0].set_title('displot - 直方图+KDE')

# 按类别分组的分布图
sns.displot(data=tips, x='total_bill', hue='time', kde=True, ax=axes[1])
axes[1].set_title('按时间分组的分布')

plt.tight_layout()
plt.show()
```

**displot优势**：
- 提供更灵活的分布可视化选项
- 支持按类别分组
- 可创建多种类型的分布图表

## 3. 双变量关系可视化

### 3.1 散点图

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 简单散点图
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=axes[0])
axes[0].set_title('简单散点图')

# 按类别着色的散点图
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size', ax=axes[1])
axes[1].set_title('按时间着色的散点图')

plt.tight_layout()
plt.show()
```

**scatterplot参数**：
- `hue`：按类别着色
- `size`：按数值大小调整点的大小
- `style`：按类别调整点的样式
- `alpha`：设置透明度

### 3.2 联合分布和边缘分布

```python
# 联合分布和边缘分布
g = sns.jointplot(data=tips, x='total_bill', y='tip')
g.fig.suptitle('联合分布和边缘分布', y=1.02)
plt.show()

# 带回归直线的联合分布图
g = sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
g.fig.suptitle('带回归直线的联合分布', y=1.02)
plt.show()

# 六边形分箱的联合分布图
g = sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
g.fig.suptitle('六边形分箱的联合分布', y=1.02)
plt.show()
```

**jointplot参数**：
- `kind`：图表类型（'scatter', 'reg', 'hex', 'kde'）
- `hue`：按类别着色
- `height`：图表大小
- `space`：边缘图与联合图之间的间距

### 3.3 成对关系图

```python
# 成对关系图
g = sns.pairplot(data=tips, vars=['total_bill', 'tip', 'size'])
g.fig.suptitle('成对关系图', y=1.02)
plt.show()

# 按类别着色的成对关系图
g = sns.pairplot(data=tips, vars=['total_bill', 'tip', 'size'], hue='time')
g.fig.suptitle('按时间着色的成对关系图', y=1.02)
plt.show()
```

**pairplot优势**：
- 一次性展示多个变量之间的关系
- 对角线显示单变量分布
- 支持按类别着色

## 4. 分类数据可视化

### 4.1 分类散点图

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 分类散点图 - swarmplot
sns.swarmplot(data=tips, x='day', y='total_bill', ax=axes[0])
axes[0].set_title('swarmplot - 分类散点图')

# 分类散点图 - stripplot
sns.stripplot(data=tips, x='day', y='total_bill', jitter=True, ax=axes[1])
axes[1].set_title('stripplot - 分类散点图')

plt.tight_layout()
plt.show()
```

**swarmplot vs stripplot**：
- `swarmplot`：点不重叠，更清晰但计算较慢
- `stripplot`：点可能重叠，使用jitter参数添加随机偏移

### 4.2 分类分布图

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 分类箱线图
sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0])
axes[0].set_title('boxplot - 分类箱线图')

# 分类小提琴图
sns.violinplot(data=tips, x='day', y='total_bill', ax=axes[1])
axes[1].set_title('violinplot - 分类小提琴图')

plt.tight_layout()
plt.show()
```

**箱线图 vs 小提琴图**：
- 箱线图：展示统计摘要（四分位数、中位数、异常值）
- 小提琴图：结合了箱线图和核密度估计，展示数据分布

### 4.3 分类柱状图

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 分类计数图
sns.countplot(data=tips, x='day', ax=axes[0])
axes[0].set_title('countplot - 分类计数图')

# 分组计数图
sns.countplot(data=tips, x='day', hue='time', ax=axes[1])
axes[1].set_title('分组计数图')

plt.tight_layout()
plt.show()

# 分类条形图（显示平均值）
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill', hue='time', ax=ax)
ax.set_title('barplot - 分类条形图')
plt.show()
```

**countplot vs barplot**：
- `countplot`：显示每个类别的计数
- `barplot`：默认显示每个类别的平均值，可通过estimator参数修改

## 5. 线性回归可视化

### 5.1 简单线性回归

```python
# 简单线性回归
fig, ax = plt.subplots(figsize=(10, 6))
sns.regplot(data=tips, x='total_bill', y='tip', ax=ax)
ax.set_title('简单线性回归')
plt.show()

# 带置信区间的线性回归
fig, ax = plt.subplots(figsize=(10, 6))
sns.regplot(data=tips, x='total_bill', y='tip', ci=95, ax=ax)
ax.set_title('带95%置信区间的线性回归')
plt.show()
```

**regplot参数**：
- `ci`：置信区间（默认95）
- `scatter_kws`：散点图参数
- `line_kws`：回归线参数
- `order`：多项式回归阶数

### 5.2 分组线性回归

```python
# 按类别分组的线性回归
fig, ax = plt.subplots(figsize=(10, 6))
sns.lmplot(data=tips, x='total_bill', y='tip', hue='time', ax=ax)
ax.set_title('按时间分组的线性回归')
plt.show()

# 多子图线性回归
g = sns.lmplot(data=tips, x='total_bill', y='tip', col='day', col_wrap=2, height=4)
g.fig.suptitle('按天分组的线性回归', y=1.02)
plt.tight_layout()
plt.show()
```

**lmplot优势**：
- 支持按类别分组绘制多个回归线
- 可创建多子图布局
- 自动处理图例

## 6. 多变量数据可视化

### 6.1 多面板网格图

```python
# 使用FacetGrid创建多面板网格图
g = sns.FacetGrid(tips, col='time', row='sex')
g.map(sns.histplot, 'total_bill')
g.fig.suptitle('按时间和性别分组的消费分布', y=1.02)
plt.tight_layout()
plt.show()

# 自定义映射函数
g = sns.FacetGrid(tips, col='day', height=4)
g.map(sns.scatterplot, 'total_bill', 'tip')
g.fig.suptitle('按天分组的消费与小费关系', y=1.02)
plt.tight_layout()
plt.show()
```

**FacetGrid用法**：
- `col`：按列分组
- `row`：按行分组
- `hue`：按颜色分组
- `map()`：将绘图函数应用到每个子图

### 6.2 分类网格图

```python
# 使用CatPlot创建分类网格图
g = sns.catplot(data=tips, x='day', y='total_bill', hue='time', col='sex', kind='box')
g.fig.suptitle('按性别和时间分组的消费分布', y=1.02)
plt.tight_layout()
plt.show()

# 不同类型的分类网格图
g = sns.catplot(data=tips, x='day', y='total_bill', hue='time', col='sex', kind='violin')
g.fig.suptitle('按性别和时间分组的消费分布（小提琴图）', y=1.02)
plt.tight_layout()
plt.show()
```

**catplot参数**：
- `kind`：图表类型（'strip', 'swarm', 'box', 'violin', 'boxen', 'point', 'bar', 'count'）
- `col`：按列分组
- `row`：按行分组
- `hue`：按颜色分组

## 7. 热图与相关性分析

### 7.1 相关性热图

```python
# 计算相关性矩阵
corr = tips.corr()

# 绘制相关性热图
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', square=True, linewidths=0.5, ax=ax)
ax.set_title('相关性热图')
plt.show()

# 加载更大的数据集
iris = sns.load_dataset('iris')
corr_iris = iris.corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr_iris, annot=True, cmap='viridis', square=True, linewidths=0.5, ax=ax)
ax.set_title('鸢尾花数据集相关性热图')
plt.show()
```

**heatmap参数**：
- `data`：二维数据（如相关性矩阵）
- `annot`：是否显示数值
- `cmap`：颜色映射
- `square`：是否绘制为正方形
- `linewidths`：网格线宽度

### 7.2 聚类热图

```python
# 绘制聚类热图
fig, ax = plt.subplots(figsize=(10, 8))
sns.clustermap(corr_iris, cmap='viridis', annot=True, figsize=(10, 8))
plt.suptitle('鸢尾花数据集聚类热图', y=1.02)
plt.tight_layout()
plt.show()
```

**clustermap特点**：
- 自动对行和列进行聚类
- 可视化数据的相似性
- 适合分析多维数据的结构

## 8. 时间序列数据可视化

### 8.1 时间序列折线图

```python
# 创建时间序列数据
date_range = pd.date_range('2024-01-01', '2024-01-31', freq='D')
data = pd.DataFrame({    'date': date_range,
    'value': np.sin(np.arange(len(date_range))/3) * 10 + np.random.randn(len(date_range)) * 2,
    'category': np.random.choice(['A', 'B', 'C'], len(date_range))
}).set_index('date')

# 绘制时间序列折线图
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(data=data, x='date', y='value', hue='category', ax=ax)
ax.set_title('时间序列数据可视化')
ax.set_xlabel('日期')
ax.set_ylabel('值')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**lineplot时间序列特点**：
- 自动处理时间索引
- 支持按类别着色
- 适合展示时间趋势

### 8.2 时间序列热力图

```python
# 创建更复杂的时间序列数据
dates = pd.date_range('2024-01-01', '2024-01-31', freq='D')
hours = range(24)

# 创建数据透视表
data = []
for date in dates:
    for hour in hours:
        value = np.sin(hour/4) * 5 + np.cos(pd.to_datetime(date).day/5) * 3 + np.random.randn()
        data.append({'date': date, 'hour': hour, 'value': value})

df = pd.DataFrame(data)
pivot = df.pivot('hour', 'date', 'value')

# 绘制时间序列热力图
fig, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(pivot, cmap='viridis', ax=ax)
ax.set_title('小时-日期热力图')
ax.set_xlabel('日期')
ax.set_ylabel('小时')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**时间序列热力图**：
- 适合展示二维时间数据
- 颜色深浅表示数值大小
- 可发现时间模式

## 9. 高级技巧与定制

### 9.1 自定义图表样式

```python
# 自定义图表样式
fig, ax = plt.subplots(figsize=(10, 6))

# 自定义散点图
scatter = sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    sizes=(50, 200),
    palette=['#ff7f0e', '#1f77b4'],
    style='time',
    markers=['o', 's'],
    alpha=0.7,
    ax=ax
)

# 自定义标题和标签
ax.set_title('消费与小费关系', fontsize=16, fontweight='bold')
ax.set_xlabel('消费金额', fontsize=12)
ax.set_ylabel('小费金额', fontsize=12)

# 自定义图例
legend = ax.legend(title='时间', loc='upper left')
legend.get_title().set_fontsize(12)

# 添加网格
ax.grid(True, alpha=0.3)

plt.show()
```

**自定义技巧**：
- `palette`：自定义颜色
- `style`：自定义标记样式
- `sizes`：自定义点大小范围
- 直接使用Matplotlib的方法进行进一步定制

### 9.2 结合Matplotlib

```python
# 结合Matplotlib进行定制
fig, ax = plt.subplots(figsize=(12, 6))

# 使用Seaborn绘制基础图表
sns.boxplot(data=tips, x='day', y='total_bill', ax=ax)

# 使用Matplotlib添加额外元素
ax.set_title('每日消费分布', fontsize=16)
ax.set_xlabel('星期', fontsize=12)
ax.set_ylabel('消费金额', fontsize=12)

# 添加水平参考线
mean_value = tips['total_bill'].mean()
ax.axhline(mean_value, color='red', linestyle='--', label=f'平均值: {mean_value:.2f}')

# 添加文本注释
ax.text(3.5, mean_value + 2, f'平均值: {mean_value:.2f}', color='red')

ax.legend()
plt.show()
```

**Seaborn与Matplotlib结合**：
- Seaborn基于Matplotlib，可直接使用Matplotlib的方法
- 先使用Seaborn创建基础图表
- 再使用Matplotlib进行精细定制

### 9.3 保存图表

```python
# 保存图表
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', ax=ax)
ax.set_title('消费与小费关系')

# 保存为不同格式
fig.savefig('seaborn_scatter.png', dpi=300, bbox_inches='tight')
fig.savefig('seaborn_scatter.pdf', bbox_inches='tight')
fig.savefig('seaborn_scatter.svg', bbox_inches='tight')

print('图表已保存为PNG、PDF和SVG格式')
```

**保存图表参数**：
- `dpi`：分辨率
- `bbox_inches='tight'`：裁剪白边
- 支持格式：PNG、PDF、SVG等
