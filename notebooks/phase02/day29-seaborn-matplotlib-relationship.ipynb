# Seaborn 与 Matplotlib 关系详解

## 核心问题

1. Seaborn 绘制图表时，如何知道绘制在 Matplotlib 创建的画布上？
2. Seaborn 是否可以不依赖 Matplotlib 单独使用？

## 1. Seaborn 如何知道绘制在 plt 创建的画布上

### 工作原理：Matplotlib 状态机

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 示例：Seaborn 如何使用当前活动的画布
print('=== Seaborn 与 Matplotlib 交互示例 ===')

# 1. 创建一个 Matplotlib 画布
plt.figure(figsize=(10, 6))
print('创建了 Matplotlib 画布')

# 2. Seaborn 自动使用这个画布
data = np.random.randn(100)
sns.histplot(data)
print('Seaborn 在当前画布上绘制')

# 3. Matplotlib 操作作用于同一画布
plt.title('Seaborn 在 Matplotlib 画布上绘制')
print('Matplotlib 设置标题')

# 4. 保存图表
plt.savefig('test_figure.png')
print('保存图表')

plt.close()
print('关闭画布')
```

### 详细机制

1. **Matplotlib 状态管理**：
   - Matplotlib 维护一个全局状态，跟踪当前活动的 Figure 和 Axes
   - 当调用 `plt.figure()` 时，会创建一个新的 Figure 并将其设置为当前活动 Figure
   - 所有后续的 `plt.` 操作都会作用于这个当前活动的 Figure

2. **Seaborn 的自动检测**：
   - 当 Seaborn 函数被调用时，它会检查是否有当前活动的 Axes
   - 如果存在活动 Axes，Seaborn 会在该 Axes 上绘制
   - 如果不存在活动 Axes，Seaborn 会自动创建一个新的 Figure 和 Axes

3. **对象层次**：
   - Figure（画布）包含一个或多个 Axes（坐标系）
    
   - Seaborn 主要操作 Axes 对象
    
   - Matplotlib 负责 Figure 的管理（创建、保存、关闭）

## 2. Seaborn 是否可以不依赖 Matplotlib 单独使用

```python
# 测试：尝试不导入 Matplotlib 直接使用 Seaborn
print('=== 测试 Seaborn 是否可以独立使用 ===')

# 只导入 Seaborn
import seaborn as sns
import numpy as np

# 创建数据
data = np.random.randn(100)

try:
    # 尝试绘制图表
    sns.histplot(data)
    print('Seaborn 可以独立使用')
except Exception as e:
    print(f'错误: {e}')
    print('Seaborn 依赖 Matplotlib')

# 检查 Seaborn 的依赖
print('=== Seaborn 依赖检查 ===')
print(f'Seaborn 版本: {sns.__version__}')
print('Seaborn 内部导入了 Matplotlib')
```

### 结论：Seaborn 不能独立使用

1. **技术依赖**：
   - Seaborn 是 Matplotlib 的高级封装库
   - Seaborn 内部导入并使用 Matplotlib 的功能
   - 所有 Seaborn 图表最终都通过 Matplotlib 渲染

2. **对象关系**：
   - Seaborn 创建的图表是 Matplotlib 的 Figure 和 Axes 对象
   - Seaborn 只是提供了更高级的 API 接口
   - 底层绘图引擎仍然是 Matplotlib

3. **安装依赖**：
   - 安装 Seaborn 时会自动安装 Matplotlib
   - 单独安装 Seaborn 而不安装 Matplotlib 是不可能的

## 3. 最佳实践：显式指定 Axes

```python
# 推荐：显式指定 Axes
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. 显式创建 Figure 和 Axes
fig, ax = plt.subplots(figsize=(10, 6))

# 2. 在指定的 Axes 上绘制
data = np.random.randn(100)
sns.histplot(data, ax=ax)

# 3. 使用 Axes 对象设置属性
ax.set_title('显式指定 Axes 示例')
ax.set_xlabel('值')
ax.set_ylabel('频率')

# 4. 保存图表
fig.savefig('explicit_axes_example.png')
print('图表已保存')

plt.close()
```

### 显式指定 Axes 的优势

1. **清晰可控**：明确指定图表的绘制位置
2. **避免混淆**：不依赖全局状态，减少意外行为
3. **多子图支持**：便于创建复杂的多子图布局
4. **代码可读性**：更清晰地表达绘图意图

## 4. 核心关系总结

| 特性 | Matplotlib | Seaborn |
|------|------------|---------|
| **角色** | 底层绘图引擎 | 高级统计可视化接口 |
| **依赖关系** | 独立库 | 依赖 Matplotlib |
| **绘制方式** | 直接控制 | 自动使用当前 Axes 或创建新的 |
| **API 风格** | 命令式 + 面向对象 | 声明式，更简洁 |
| **图表对象** | Figure, Axes | 基于 Matplotlib 对象 |
| **主要优势** | 高度可定制，底层控制 | 统计功能丰富，默认样式美观 |

2. 对于简单图表：可以使用默认的状态机机制

## 6. 常见问题解答

### Q1: 为什么 Seaborn 代码中看不到 Matplotlib 的使用？
**A1**: Seaborn 隐藏了底层的 Matplotlib 调用，提供了更简洁的 API。但实际上，Seaborn 在内部使用 Matplotlib 的功能。

### Q2: Seaborn 和 Matplotlib 可以同时使用吗？
**A2**: 当然可以，而且推荐这样做。使用 Seaborn 创建美观的基础图表，使用 Matplotlib 进行精细定制。

### Q3: 如何在一个 Figure 中创建多个 Seaborn 图表？
**A3**: 使用 `plt.subplots()` 创建多个 Axes，然后在每个 Axes 上调用 Seaborn 函数。

## 7. 总结

Seaborn 和 Matplotlib 是 Python 数据可视化的黄金组合：

1. **Seaborn** 提供了高级统计可视化功能，代码简洁，默认样式美观
2. **Matplotlib** 提供了底层控制能力，高度可定制，支持复杂图表
3. **两者结合** 可以发挥各自的优势，创建既美观又功能强大的数据可视化

Seaborn 不能独立使用，它依赖 Matplotlib 作为底层绘图引擎。理解它们之间的关系，掌握正确的使用方法，可以让你在数据可视化方面更加得心应手。
